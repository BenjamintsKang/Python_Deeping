```python
from random import *

class Account:
    account = 0

    def __init__(self, name, money):
        self.name = name            # 이름
        self.depositMoney = money   # 초기 잔액
        self.bank = 'SC은행'        #은행명

        num1 = randint(0, 999)
        num2 = randint(0, 99)
        num3 = randint(0, 999999)

        num1 = str(num1).zfill(3)
        num2 = str(num2).zfill(2)
        num3 = str(num3).zfill(6)

        self.banknum = num1+'-'+ num2+'-'+num3  # 은행 계좌 랜덤 생성

        Account.account += 1   # 계좌 생성 횟수
        self.depositCount = 0   # 입금 횟수

        self.depositlog = []
        self.withdrawlog = []

    def get_account_num(self):
        print('생성된 계좌의 갯수는 {0} 개 입니다.'.format(self.account))

    def deposit(self, money):

        if money > 0:
            self.depositMoney += money
            self.depositCount += 1 # 입금 횟수 + 1
            self.depositlog.append(money)
            print('{0} 님의 계좌에 {1} 원이 입금되었습니다. 총 금액은 {2} 입니다.'.format(self.name, format(money, ','), format(self.depositMoney, ',')))

            if self.depositCount % 5 == 0: # 입금 횟수가 5회가 될 때
                self.depositMoney *= 1.01 # 1% 이자 추가
                print('입금 횟수가 {0} 회 이므로 1%의 이자가 붙습니다. 총 금액은 {1} 입니다.'.format(self.depositCount, format(self.depositMoney, ',')))
        else:
            print('최소 입금 금액은 1원 입니다.')

    def withdraw(self, money):
        if self.depositMoney-money > 0:
            self.depositMoney -= money
            self.withdrawlog.append(money)
            print('{0} 원이 출금되었습니다. 남은 금액은 {1} 입니다.'.format(format(money, ','), format(self.depositMoney, ',')))
        else:
            print('예치금보다 더 많은 돈을 출금할 수 없습니다.')

    def display_info(self):
        print('은행이름 : {0}, 예금주 : {1}, 계좌번호 : {2}, 잔고 : {3}'.format(self.bank, self.name, self.banknum, format(self.depositMoney, ',')))


    def deposit_history(self):
        print('{0} 님의 입금 기록은 : {1} 입니다.'.format(self.name, self.depositlog))

    def withdraw_history(self):
        print('{0} 님의 출금 기록은 : {1} 입니다.'.format(self.name, self.withdrawlog))

namelist = []

a = Account('유제민', 10000)
b = Account('홍길동', 30000)
c = Account('도우너', 5000)

namelist.append(a)
namelist.append(b)
namelist.append(c)

a.get_account_num()

a.deposit(1000)
b.deposit(1500000)
c.deposit(3000)

a.deposit(500)
a.deposit(3000)
a.deposit(2000)
a.deposit(100)
a.deposit(2000000)

a.withdraw(500)
a.withdraw(3000)
a.withdraw(100)
a.withdraw(400)
a.withdraw(1000)

a.display_info()

for d in namelist:
    if d.depositMoney >= 100000:
        d.display_info()


a.deposit_history()
a.withdraw_history()


```
